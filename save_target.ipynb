```python
from PIL import Image
import numpy as np
import os
```

```python
def transform_pixel(pixel):
    if pixel == 1:
        return 0
    elif pixel == 2:
        return 127
    elif pixel == 3:
        return 255
    else:
        return pixel
```

```python
folder_path = "data/oxford-iiit-pet/annotations/trimaps"

def process_target(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp')) and not filename.startswith("."):  # 対応する画像フォーマット
            image_path = os.path.join(folder_path, filename)
            image = Image.open(image_path)
            #image = image.convert('L')  # グレースケール画像に変換
            pixel_array = np.array(image)
            
            # 画素値を変換
            vectorized_transform = np.vectorize(transform_pixel)
            transformed_array = vectorized_transform(pixel_array)
            
            # 新しい画像として保存
            transformed_image = Image.fromarray(np.uint8(transformed_array))
            new_filename = f"transformed_{filename}"
            new_folder_path = os.path.join(folder_path, "target")
            os.makedirs(new_folder_path, exist_ok=True)
            new_image_path = os.path.join(new_folder_path, new_filename)
            transformed_image.save(new_image_path)
            print(f"Saved transformed image: {new_image_path}")

process_target(folder_path)
```
